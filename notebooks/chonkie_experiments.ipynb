```python
from chonkie import MarkdownChef

chef = MarkdownChef(tokenizer="o200k_harmony",)

doc = chef.process("D:\Pdf\KrishiMitra (කෘෂි මිතුරා)_ Agentic Farm Advisory S.md")
```

```python
# Step 1: Extract ALL component boundaries from MarkdownChef output
# This creates a sorted list of positions where components start/end

def extract_boundaries(doc):
    """
    Extract all boundaries from tables, code blocks, and images.
    Returns sorted list of unique positions.
    """
    boundaries = [0, len(doc.content)]  # Start and end of document
    
    # Add table boundaries
    for table in doc.tables:
        boundaries.append(table.start_index)
        boundaries.append(table.end_index)
    
    # Add code block boundaries
    for code in doc.code:
        boundaries.append(code.start_index)
        boundaries.append(code.end_index)
    
    # Note: Images will be replaced with descriptions in doc.content
    # So we don't need their boundaries for segmentation
    
    # Sort and remove duplicates
    boundaries = sorted(set(boundaries))
    
    return boundaries

# Test with our document
boundaries = extract_boundaries(doc)
print(f"Found {len(boundaries)} boundaries:")
print(boundaries[:10])  # Show first 10
```

```python
# Step 2: Create O(1) lookup for components by start position
# This allows fast checking: "Is position X the start of a component?"

def build_component_map(doc):
    """
    Build a dictionary mapping start_index → component object.
    Allows O(1) lookup to check if a boundary is a component start.
    """
    component_map = {}
    
    # Map tables
    for table in doc.tables:
        component_map[table.start_index] = {
            "type": "table",
            "start": table.start_index,
            "end": table.end_index,
            "content": table.content,
            "obj": table
        }
    
    # Map code blocks
    for code in doc.code:
        component_map[code.start_index] = {
            "type": "code",
            "start": code.start_index,
            "end": code.end_index,
            "content": code.content,
            "language": code.language,
            "obj": code
        }
    
    return component_map

# Build the map
comp_map = build_component_map(doc)
print(f"Component map has {len(comp_map)} entries")
print("\nExample component:")
if comp_map:
    first_key = list(comp_map.keys())[0]
    print(f"Position {first_key}: {comp_map[first_key]['type']}")
```

```python
# Step 3: Build ordered segments from boundaries
# Walk through boundaries in order, labeling each segment as text/table/code

def build_segments(boundaries, component_map, doc_content):
    """
    Build ordered list of segments from position 0 to end.
    Each segment is either text, table, or code.
    
    Algorithm:
    - For each consecutive pair of boundaries (start, end):
      - Check if start position is in component_map
      - If yes: segment is that component
      - If no: segment is text
    """
    segments = []
    
    for i in range(len(boundaries) - 1):
        start = boundaries[i]
        end = boundaries[i + 1]
        
        # Check if this position starts a component
        if start in component_map:
            # This segment is a table/code component
            comp = component_map[start]
            segments.append({
                "type": comp["type"],
                "start": start,
                "end": end,
                "content": comp["content"],
                "obj": comp.get("obj"),
                "language": comp.get("language")  # For code blocks
            })
        else:
            # This segment is plain text
            segments.append({
                "type": "text",
                "start": start,
                "end": end,
                "content": doc_content[start:end]
            })
    
    return segments

# Build segments
segments = build_segments(boundaries, comp_map, doc.content)
print(f"Created {len(segments)} segments\n")

# Show first 5 segments in order
for i, seg in enumerate(segments[:5]):
    print(f"Segment {i}: {seg['type']:6s} | Pos {seg['start']:5d}-{seg['end']:5d} | Length: {seg['end']-seg['start']:4d} chars")
```

```python
# Step 4: Chunk each segment with appropriate chunker
# Process segments IN ORDER from position 0 to end

from chonkie import SentenceChunker, TableChunker, CodeChunker
import tiktoken

def chunk_segments_in_order(segments, child_chunk_size=800, child_overlap=100):
    """
    Chunk each segment with the appropriate chunker.
    Maintains processing order: 0 → end.
    
    Returns list of chunks in document order.
    """
    all_chunks = []
    
    # Initialize chunkers
    text_chunker = SentenceChunker(
        tokenizer="o200k_harmony",
        chunk_size=child_chunk_size,
        chunk_overlap=child_overlap,
        min_sentences_per_chunk=1
    )
    
    table_chunker = TableChunker(
        tokenizer="row",  # Chunk by rows
        chunk_size=50     # Max rows per chunk
    )
    
    # Process each segment in order
    for seg_idx, segment in enumerate(segments):
        seg_type = segment["type"]
        content = segment["content"]
        
        # Skip empty segments
        if not content or not content.strip():
            continue
        
        # Chunk based on type
        if seg_type == "text":
            chunks = text_chunker.chunk(content)
            for chunk in chunks:
                all_chunks.append({
                    "type": "text",
                    "text": chunk.text,
                    "token_count": chunk.token_count,
                    "original_start": segment["start"] + chunk.start_index,
                    "original_end": segment["start"] + chunk.end_index,
                    "segment_id": seg_idx
                })
        
        elif seg_type == "table":
            # TableChunker preserves table structure (header + rows)
            chunks = table_chunker.chunk(content)
            for chunk in chunks:
                all_chunks.append({
                    "type": "table",
                    "text": chunk.text,
                    "token_count": chunk.token_count,
                    "original_start": segment["start"],
                    "original_end": segment["end"],
                    "segment_id": seg_idx
                })
        
        elif seg_type == "code":
            # For code, use CodeChunker if available, else keep whole
            # Check if code is small enough to keep whole
            tokenizer = tiktoken.get_encoding("o200k_harmony")
            code_tokens = len(tokenizer.encode(content))
            
            if code_tokens <= child_chunk_size:
                # Keep whole code block
                all_chunks.append({
                    "type": "code",
                    "text": content,
                    "token_count": code_tokens,
                    "language": segment.get("language"),
                    "original_start": segment["start"],
                    "original_end": segment["end"],
                    "segment_id": seg_idx,
                    "is_complete": True
                })
            else:
                # Large code - use CodeChunker
                code_chunker = CodeChunker(
                    language=segment.get("language", "python"),
                    tokenizer="o200k_harmony",
                    chunk_size=child_chunk_size,
                    include_nodes=False
                )
                chunks = code_chunker.chunk(content)
                for chunk in chunks:
                    all_chunks.append({
                        "type": "code",
                        "text": chunk.text,
                        "token_count": chunk.token_count,
                        "language": segment.get("language"),
                        "original_start": segment["start"],
                        "original_end": segment["end"],
                        "segment_id": seg_idx,
                        "is_complete": False  # Partial code block
                    })
    
    return all_chunks

# Chunk all segments
final_chunks = chunk_segments_in_order(segments, child_chunk_size=800, child_overlap=100)

print(f"Total chunks created: {len(final_chunks)}\n")
print("First 10 chunks:")
for i, chunk in enumerate(final_chunks[:10]):
    print(f"{i+1}. Type: {chunk['type']:6s} | Tokens: {chunk['token_count']:4d} | Text preview: {chunk['text'][:60]}...")
```

```python
# Verify order preservation: chunks should be in position order

# Check that chunks are ordered by original_start
prev_start = -1
out_of_order = 0

for i, chunk in enumerate(final_chunks):
    current_start = chunk["original_start"]
    if current_start < prev_start:
        out_of_order += 1
        print(f"⚠️ Chunk {i} out of order: prev={prev_start}, current={current_start}")
    prev_start = current_start

if out_of_order == 0:
    print("✅ All chunks are in document order (position 0 → end)")
else:
    print(f"⚠️ Found {out_of_order} chunks out of order")

# Show chunk type distribution
from collections import Counter
type_counts = Counter(chunk["type"] for chunk in final_chunks)
print(f"\nChunk types: {dict(type_counts)}")
```
