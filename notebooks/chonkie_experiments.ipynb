```python
import fitz

pdf_path = r"D:\Downloads\Crime_and_Punishment_T.pdf"
doc = fitz.open(pdf_path)

text = ""
for page_num in range(len(doc)):
    page = doc[page_num]
    text += page.get_text()

doc.close()
print(text)
```

```python
import tiktoken
tokenizer = tiktoken.get_encoding("o200k_harmony")
tokens = tokenizer.encode(text)
print(f"Total tokens: {len(tokens)}")
```

```python
from chonkie import SentenceChunker

chunker = SentenceChunker(
    tokenizer="o200k_harmony",     
    chunk_size= 30000,          
    chunk_overlap=5000,         
    min_sentences_per_chunk=1  
)
```

```python
chunks = chunker.chunk(text)
```

```python
from chonkie import SentenceChunker

High_level_chunker = SentenceChunker(
    tokenizer="o200k_harmony",     
    chunk_size= 30000,          
    chunk_overlap=5000,         
    min_sentences_per_chunk=1  
)

low_level_chunker = SentenceChunker(
    tokenizer="o200k_harmony",     
    chunk_size= 800,          
    chunk_overlap=100,       
    min_sentences_per_chunk=1 
)
```

```python
context_chunks = High_level_chunker.chunk(text)
context_chunks_text = [chunk.text for chunk in context_chunks]
full_chunks = low_level_chunker.chunk_batch(context_chunks_text)
```
